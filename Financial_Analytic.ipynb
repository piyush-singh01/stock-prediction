## Importing the Required Libraries

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
import itertools

import statsmodels.api as sm
import statsmodels
```

```python
df = pd.read_csv('nifty.csv')
df
```

```python
df['Date']=pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.set_index(['Date'], inplace=True)
df2 = df  ## Creating a copy for later use
df
```

```python
df = df['Close']
df
```

## Plotting the value of index over time (1 Year Period)

```python
df.plot()
plt.ylabel('Closing')
plt.xlabel('Date')
plt.show()
```

```python
# Creating tuples of all possible combinations of p, d, q in range 0 to 4 (inclusive)
p = d = q = range(0,5)
pdq = list(itertools.product(p, d, q))
print(pdq)
```

```python
train_data = df['2022-03-29':'2023-03-29']
test_data = df['2023-03-29':'2023-03-29']
```

```python
display(train_data)
```

```python
# Looping over all possible tuples of p,d,q generated previously to find AIC values of models fitted using all possible combinations
warnings.filterwarnings("ignore")
AIC = []
ARIMA_model = []
for param in pdq:
    mod = statsmodels.tsa.arima.model.ARIMA(train_data,
                                           order=param,
                                           enforce_stationarity=False,
                                           enforce_invertibility=False)
    res = mod.fit()
    print('ARIMA{} - AIC:{}'.format(param, res.aic), end='\r')
    AIC.append(res.aic)
    ARIMA_model.append([param])
```

```python
print('The smallest AIC is {} for model ARIMA{}'.format(min(AIC),ARIMA_model[AIC.index(min(AIC))][0]))
```

```python
## Fit the model with the given parameters
mod = statsmodels.tsa.arima.model.ARIMA(train_data,
                                order=ARIMA_model[AIC.index(min(AIC))][0],
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()
```

```python
results.summary()
```

## Checking using augmented dickey fuller test for stationarity

```python
from statsmodels.tsa.stattools import adfuller
```

```python
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
## df2 is a copy of the original dataframe
## Checking for no lag
adfuller_test(df2['Close'])
```

```python
## For second difference
df2['sdiff'] = df2['Close'] - df2['Close'].shift(2)
df2
```

```python
df2['Close'].shift(2)
```

### Plotting the second difference lag

```python
df2['sdiff'].plot()
plt.ylabel('Second Difference Closing')
plt.xlabel('Date')
plt.show()
```

```python
adfuller_test(df2['sdiff'].dropna())
```

Here, for second degree lag, the p-value is much lower than the significance level. Hence, second difference is stationary.
